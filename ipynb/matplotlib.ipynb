# Matplotlib
## Functional Approach

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,11)
fx = 2 * x + 1
f2x = x**2
```

### Quick visualization

```python
plt.plot(x,fx)
plt.title("String Title")
plt.xlabel("x Axis")
plt.xlim(0,10)
plt.ylabel("y Axis")
plt.ylim(0,90)
plt.savefig("matplotlib-first-plot.png")
plt.show()
```

## OOP API

```python
# blank canvas object
fig = plt.figure(figsize=[2,2])
# add axes to the object
# position of axes
# 0,0 is lower left corner
# 1,1 is top right corner
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.5,0.5,1,1])
ax1.plot(x,fx)
ax2.plot(x,f2x)
plt.show()
```

```python
# figsize in inch
fig = plt.figure(dpi=200, figsize=(15,5))
ax1 = fig.add_axes([0,0,1,1])
ax1.set_xlabel("x Axis")
ax1.set_ylabel("y Axis")
ax1.set_title("Title")
ax1.set_xlim(0, x.max())
ax1.set_ylim(0, fx.max())
ax1.plot(x, fx)
fig.savefig("matplotlib-oop-figure.png", bbox_inches = "tight")
plt.show()
```

### Subplots

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
print(type(axes), axes.shape)
axes[0].plot(x,fx)
axes[1].plot(x,f2x)
fig.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
print(type(axes), axes.shape)
fig.suptitle("Figure title", fontsize=12)
axes[0][0].set_xlabel("x Axis")
axes[0][0].set_ylabel("y Axis")
axes[0][0].set_title("Title 0,0", fontsize=8)
axes[0][0].plot(x,fx)
axes[1][1].plot(x,f2x)
plt.subplots_adjust(wspace=0.25, hspace=0.3)
plt.show()
```

### Styling

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(
    x,
    fx,
    linewidth=3,
    linestyle=":",
    label="linear",
    color="blue"
)
ax.plot(
    x,
    f2x,
    label="quadratic",
    color="red",
    marker="o",
    markersize=5
)
ax.legend(loc=2)
plt.show()
```
