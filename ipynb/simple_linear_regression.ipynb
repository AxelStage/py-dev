# Simple Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Advertising Dataset
Shows how much money is spend for advertising on different media and the total sales amount for each week.  

```python
filepath = "../ml/DATA/Advertising.csv"
df = pd.read_csv(filepath)
```

### Engineered Feature
total_spend was created from original data

```python
df["total_spend"] = df["TV"] + df["radio"] + df["newspaper"]
```

```python
df.head()
```

### Scatterplot

```python
sns.scatterplot(data=df, x="total_spend", y="sales")
```

### Scatterplot with Simple Linear Regression

```python
sns.regplot(data=df, x="total_spend", y="sales")
```

### Convention
X (capital) for feature matrix  
y for label

```python
X = df["total_spend"]
y = df["sales"]
```

```python
# y = mx + b
# y = B1x + B0
# polyfit returns the beta coefficients B1 and B0
B1, B0 = np.polyfit(X,y,deg=1)
```

```python
potential_spent = np.linspace(0, 500, 100)
predicted_sales = B1 * potential_spent + B0
```

```python
sns.scatterplot(data=df, x="total_spend", y="sales")
plt.plot(potential_spent, predicted_sales, color="red")
```

### Measure Performance for Simple Linear Regression
Coefficients get small if they have less importance, see B3 and B2

```python
# y = B3x**3 + B2x**2 + B1x + B0
B3, B2, B1, B0 = np.polyfit(X,y, deg=3)
print(B3, B2, B1, B0)
```

```python
predicted_sales_2 = B3 * potential_spent**3 + B2 * potential_spent**2 + B1 * potential_spent**1 + B0
```

```python
sns.scatterplot(data=df, x="total_spend", y="sales")
plt.plot(potential_spent, predicted_sales_2, color="red")
```
