# Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../ml/05-Seaborn/dm_office_sales.csv")
```

```python
df.head()
```

## Scatter Plot

```python
plt.figure(dpi=200,figsize=(15, 5))
sns.scatterplot(
    data=df,
    x="salary",
    y="sales",
    hue="level of education",
    palette="Dark2",
    s=20, # point size
    alpha=0.7 # transparency
)
plt.show()
```

## Distribution Plots

### Rug plot

```python
plt.figure(dpi=200,figsize=(15, 5))
sns.rugplot(data=df, x="salary", height=0.9)
plt.show()
```

### Histogram

```python
sns.set(style="darkgrid") # set global!
sns.displot(
    data=df,
    x="salary",
    bins=20,
    color="red",
    edgecolor="black",
    linewidth=2
)
plt.show()
```

```python
plt.figure(dpi=200,figsize=(15, 5))
sns.histplot(
    data=df,
    x="salary",
    bins=20,
    color="red",
    edgecolor="black",
    linewidth=1,
    kde=True
)
plt.show()
```

### KDE Plot

```python
plt.figure(dpi=200,figsize=(15, 5))
sns.kdeplot(
    data=df,
    x="salary"
)
plt.show()
```

## Categorical Plots
### Bar plot

```python
plt.figure(dpi=200,figsize=(15, 5))
sns.countplot(
    data=df,
    x="level of education",
    hue="division",
    palette="Set2"
)
plt.show()
```

```python
plt.figure(dpi=200,figsize=(15, 5))
sns.barplot(
    data=df,
    x="level of education",
    y="salary",
    estimator=np.mean,
    hue="division"
)
plt.errorbar = 'ci', 0.95
plt.legend(bbox_to_anchor=(1, 1.02))
plt.show()
```

### Box Plot

```python
df = pd.read_csv("../ml/05-Seaborn/StudentsPerformance.csv")
```

```python
df.head()
```

```python
plt.figure(dpi=200,figsize=(15, 5))
sns.boxplot(data=df,x="test preparation course", y="math score")
plt.show()
```

```python
plt.figure(dpi=200,figsize=(15, 5))
sns.boxplot(data=df,x="parental level of education", y="math score", hue="test preparation course")
plt.show()
```

```python
plt.figure(dpi=200,figsize=(15, 5))
sns.boxplot(data=df,x="math score", y="parental level of education", hue="test preparation course")
plt.show()
```

### Violin Plot
Mirrored KDE function, turned by 90 degrees
Inside is the Mean and IQR

```python
plt.figure(dpi=200,figsize=(15, 5))
sns.violinplot(
    data=df,
    x="parental level of education",
    y="math score",
    hue="test preparation course",
    palette="Set1"
)
plt.show()
```

```python
plt.figure(dpi=200,figsize=(15,5))
sns.violinplot(
    data=df,
    x="parental level of education",
    y="math score",
    hue="test preparation course",
    palette="Set1",
    split=True,
    inner="quartile"
)
plt.show()
```

### Swarm Plot

```python
plt.figure(dpi=200, figsize=(15,5))
sns.swarmplot(
    data=df,
    x="math score",
    y="gender",
    hue="test preparation course",
    size=3,
    dodge=True,
    palette="Set1"
)
plt.show()
```

### Boxen Plot

```python
plt.figure(dpi=200, figsize=(15,5))
sns.boxenplot(
    data=df,
    x="math score",
    y="test preparation course"
)
plt.show()
```

## Comparison Plots
### Joint Plot

```python
sns.jointplot(data=df,x="math score", y="reading score", kind="scatter", alpha=0.6)
plt.show()
```

```python
sns.jointplot(data=df,x="math score", y="reading score", kind="hex")
plt.show()
```

```python
sns.jointplot(data=df,x="math score", y="reading score", kind="kde", fill=True)
plt.show()
```

```python
sns.jointplot(data=df,x="math score", y="reading score", kind="scatter", hue="gender")
plt.show()
```

### Pair Plot

```python
sns.pairplot(data=df)
plt.show()
```

```python
sns.pairplot(data=df, hue="gender", corner=True)
plt.show()
```

## Grid Plots

```python
sns.catplot(data=df, x="gender", y="math score", kind="box", col="lunch")
plt.show()
```

```python
sns.catplot(data=df, x="gender", y="math score", kind="strip", col="lunch", row="test preparation course")
plt.show()
```

```python
g = sns.PairGrid(df)
g = g.map_upper(sns.scatterplot)
g = g.map_diag(sns.histplot)
g = g.map_lower(sns.kdeplot)
```

## Matrix Plot
Columns should have same units!

```python
df = pd.read_csv("../ml/05-Seaborn/country_table.csv")
```

```python
df.head()
```

```python
df = df.set_index("Countries")
df = df.drop("Life expectancy", axis=1)
```

```python
plt.figure(dpi=200, figsize=(15,5))
sns.heatmap(df, linewidth=0.5, annot=True, cmap="viridis")
plt.show()
```

```python
plt.figure(dpi=200, figsize=(15,5))
sns.clustermap(df, linewidth=0.5, annot=True, cmap="viridis")
plt.show()
```
