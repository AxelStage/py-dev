```python
def last_four(number):
    return str(number)[-4:]
df["last_four"] = df["CC Number"].apply(last_four)
df.head()
```

```python
df["last_four2"] = df["CC Number"].apply(lambda number: str(number)[-4:])
df.head()
```

```python
def tip_quality(total_bill, tip):
    tip_ratio = tip / total_bill
    if tip_ratio > 0.25:
        return "Generous"
    else:
        return "Other"

df["Generous"] = df[["total_bill", "tip"]].apply( lambda df: tip_quality(df["total_bill"], df["tip"]), axis=1 )
df
```

```python
df["Generous2"] = np.vectorize(tip_quality)(df["total_bill"],df["tip"])
df.head()
```
